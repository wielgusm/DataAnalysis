```python
alist = hops.read_alist('/Volumes/SanDiskMW/Data2017/alist.v6.5s')
util.add_gmst(alist)
#sources = ['3C279','OJ287','J1924-2914','1055+018','1749+096','1921-293','CYGX-3','J1733-1304','CENA']
QuadL = [['AZ','XL','AL','XZ'],['AZ','XS','AS','XZ'],['AZ','XJ','AJ','XZ'], ['AZ','XR','AR','XZ'],
         ['AZ','XP','AP','XZ'],['AS','XP','AP','XS'],['AJ','XR','AR','XJ'], ['AJ','XL','AL','XJ'],
         ['AJ','XP','AP','XJ'],['AJ','XS','AS','XJ'],['AL','XP','AP','XL'], ['AL','XR','AR','XL'],
         ['AS','XR','AR','XS'],['AS','XJ','AJ','XS'],['AS','XL','AL','XS'], ['AP','XR','AR','XP'],
         ['AL','XS','AS','XL'],['AS','XR','AR','XS'],
         ['ZJ','LS','JL','ZS'],['ZJ','XS','XJ','ZS'],['ZJ','AS','AJ','ZS'],['ZJ','SR','JR','ZS'],
         ['XJ','SR','JR','XS'],['XJ','LS','JL','XS'],['AJ','SL','JL','AS'],['AJ','PS','JP','AS'],
         ['AJ','SR','AS','JR'],['ZJ','PS','ZS','JP'],['XJ','PS','XS','JP']]
```

```python
pol='LL'
for Quad in QuadL:   
    try:
        TrianglesA.plotQuadranglePolar(alist,Quad,pol,1.,method=1)
    except (IndexError,KeyError):
        print 'Error for triangle '+str(Quad)
        continue
```

```python
pol='RR'
for Quad in QuadL:   
    try:
        TrianglesA.plotQuadranglePolar(alist,Quad,pol,1.,method=1)
    except (IndexError,KeyError):
        print 'Error for triangle '+str(Quad)
        continue
```
